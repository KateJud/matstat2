1. Рассчитайте 100 псевдослучайных чисел методом серединных квадратов <br><br>
Допустим, мы хотим генерировать числа с четырьмя знаками после запятой. Для этого
берётся некое начальное четырёхзначное число z1 и возводится в квадрат — получается
восьмизначное число. Средние четыре цифры этого восьмизначного числа задают
следующий элемент этой последовательности — z2. Затем z2 возводится в квадрат и берутся
средние четыре цифры от квадрата и т. д.
В конце укладываем все полученные числа в пределы от нуля до единицы простым

```python
def random_generator(initial_number):
    """
        Генератор случайной последовательности.
        Реализован на основе метода срединных квадратов.
    """
 
    if not isinstance(initial_number, int):
        raise ValueError("Входное значение не является числом!")
 
    while True:
        square_str = str(initial_number ** 2)
        #print(square_str)
        while len(square_str)!=8:
            square_str='0'+square_str
        #print(square_str)
        start_index = len(square_str) // 4
        finish_index = start_index + 1 if len(square_str) % 2else start_index
 
        initial_number = int(square_str[start_index:-finish_index])
        yield initial_number
```

Генерируем 100 чисел методом серединных квадратов

```python
generator = random_generator(1661)
numbers=[]
for index, number in (zip(range(100), generator)):
        print("{0}: {1}".format(index, number/10000))
        numbers.append(number/10000)
        
display(numbers)
```
